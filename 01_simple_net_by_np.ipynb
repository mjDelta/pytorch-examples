```python
import numpy as np
```

```python
N,in_,h,out_=64,1000,100,10

x=np.random.randn(N,in_)
y=np.random.randn(N,out_)
print(x.shape)
print(y.shape)
```

```python
w1=np.random.randn(in_,h)
w2=np.random.randn(h,out_)
print(w1.shape)
print(w2.shape)
```

```python
lr=1e-6
for epoch in range(500):
    h=x.dot(w1)
    h_relu=np.maximum(h,0)
    pred=h_relu.dot(w2)
    
    loss=np.sum(np.square((pred-y)))
    print("Epoch %s, loss is %s"%(epoch,loss))
    
    grad_pred=2*(pred-y)
    grad_w2=h_relu.T.dot(grad_pred)
    grad_h_relu=grad_pred.dot(w2.T)
    grad_h=grad_h_relu.copy()
    grad_h[h<0]=0
    grad_w1=x.T.dot(grad_h)
    
    w1-=lr*grad_w1
    w2-=lr*grad_w2
    
```
