```python
import torch 

N,in_,h,out_=64,1000,100,10
dtype=torch.float
device=torch.device("cpu")


x=torch.randn(N,in_,dtype=dtype,device=device)
y=torch.randn(N,out_,dtype=dtype,device=device)
```

```python
model=torch.nn.Sequential(
    torch.nn.Linear(in_,h),
    torch.nn.ReLU(),
    torch.nn.Linear(h,out_))

loss_fn=torch.nn.MSELoss(size_average=False)
```

```python
lr=1e-4
optimizer=torch.optim.Adam(model.parameters(),lr=lr)
```

```python
for epoch in range(500):
    pred=model(x)
    
    loss=loss_fn(pred,y)
    print("Epoch %s, loss is %s"%(epoch,loss.item()))
    
    optimizer.zero_grad()
    
    loss.backward()
    
    optimizer.step()
```
