```python
import torch

dtype=torch.float
device=torch.device("cpu")

N,in_,h,out_=64,1000,100,10

x=torch.randn(N,in_,dtype=dtype)
y=torch.randn(N,out_,device=device)
```

```python
class myTwoLayerNet(torch.nn.Module):
    def __init__(self,D_in,H,D_out):
        super(myTwoLayerNet,self).__init__()
        self.linear1=torch.nn.Linear(D_in,H)
        self.linear2=torch.nn.Linear(H,D_out)
    def forward(self,x):
        h_relu=self.linear1(x).clamp(min=0)
        pred=self.linear2(h_relu)
        return pred
model=myTwoLayerNet(in_,h,out_)
loss_fn=torch.nn.MSELoss(size_average=False)
optimizer=torch.optim.SGD(model.parameters(),lr=1e-4)
```

```python
for epoch in range(500):
    pred=model(x)
    loss=loss_fn(pred,y)
    print("Epoch %s, loss is %s"%(epoch,loss.item()))
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
