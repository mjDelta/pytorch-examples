```python
import torch

dtype=torch.float
device=torch.device("cpu")

N,in_,h,out_=64,1000,100,10

x=torch.randn(N,in_)
y=torch.randn(N,out_)
```

```python
model=torch.nn.Sequential(
        torch.nn.Linear(in_,h),
        torch.nn.ReLU(),
        torch.nn.Linear(h,out_)
)
loss=torch.nn.MSELoss(size_average=False)
```

```python
lr=1e-4
for epoch in range(500):
    y_pred=model(x)
    
    l=loss(y_pred,y)
    print("Epoch %s, loss is %s."%(epoch,l.item()))
    
    model.zero_grad()
    
    l.backward()
    
    with torch.no_grad():
        for param in model.parameters():
            param-=lr*param.grad
    
    
```
