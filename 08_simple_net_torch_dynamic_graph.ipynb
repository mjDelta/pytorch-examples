```python
import torch
import random

class DynamicNet(torch.nn.Module):
    def __init__(self,in_,h,out_):
        super(DynamicNet,self).__init__()
        self.input_linear=torch.nn.Linear(in_,h)
        self.middle_linear=torch.nn.Linear(h,h)
        self.output_linear=torch.nn.Linear(h,out_)
    def forward(self,x):
        h_relu=self.input_linear(x).clamp(min=0)
        iters=random.randint(0,3)
        print("Hidden layers' num is %s"%iters)
        for _ in range(iters):
            h_relu=self.middle_linear(h_relu).clamp(min=0)
        pred=self.output_linear(h_relu)
        return pred
    
```

```python
N,in_,h,out_=64,1000,100,10

x=torch.randn(N,in_)
y=torch.randn(N,out_)

model=DynamicNet(in_,h,out_)

loss_fn=torch.nn.MSELoss(size_average=False)
optimizer=torch.optim.SGD(model.parameters(),lr=1e-4)

```

```python
for epoch in range(500):
    pred=model(x)
    
    loss=loss_fn(pred,y)
    print("Epoch %s, loss is %s."%(epoch,loss.item()))
    
    optimizer.zero_grad()
    
    loss.backward()
    
    optimizer.step()
    
    
```
